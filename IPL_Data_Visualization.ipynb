```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load the DataSet
data = pd.read_csv("/content/data.csv")
```

```python
# Convert 'date' column to datetime
if 'date' in data.columns:
    data['date'] = pd.to_datetime(data['date'], errors='coerce')
```

```python
data.head()
```

```python
data.tail()
```

```python
# Fill missing values for 'city' and 'winner'
data['city'] = data['city'].fillna('Unknown')
data['winner'] = data['winner'].fillna('No Result')
```

```python
# Drop 'umpire3' column if it exists and is mostly NaN
if 'umpire3' in data.columns and data['umpire3'].isna().sum() > 0.9 * len(data):
    data = data.drop(columns=['umpire3'])
```

```python
# Visualization 1: Number of Matches Played Per Season
matches_per_season = data['season'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
matches_per_season.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Number of Matches Played Per Season", fontsize=16)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Number of Matches", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualization 2: Most Successful Teams (Total Wins)
team_wins = data['winner'].value_counts()
plt.figure(figsize=(12, 6))
team_wins.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Most Successful Teams (Total Wins)", fontsize=16)
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Number of Wins", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualization 3: City Hosting the Most Matches
city_matches = data['city'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
city_matches.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Top 10 Cities Hosting the Most Matches", fontsize=16)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Matches", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualization 4: Toss Decision Trends
toss_decision = data['toss_decision'].value_counts()
plt.figure(figsize=(8, 6))
toss_decision.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue'])
plt.title("Toss Decision Trends", fontsize=16)
plt.ylabel('')
plt.show()
```

```python
# Visualisation 5. Distribution of wins by runs
win_by_runs = data['win_by_runs'][data['win_by_runs'] > 0]
plt.figure(figsize=(10, 6))
plt.hist(win_by_runs, bins=20, color='blue', edgecolor='black')
plt.title("Distribution of Wins by Runs", fontsize=16)
plt.xlabel("Runs", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualisation 6. Distribution of wins by wickets
win_by_wickets = data['win_by_wickets'][data['win_by_wickets'] > 0]
plt.figure(figsize=(10, 6))
plt.hist(win_by_wickets, bins=10, color='green', edgecolor='black')
plt.title("Distribution of Wins by Wickets", fontsize=16)
plt.xlabel("Wickets", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualisation 7. Matches played by each team
teams = pd.concat([data['team1'], data['team2']])
team_counts = teams.value_counts()
plt.figure(figsize=(10, 6))
plt.bar(team_counts.index, team_counts.values, color='purple')
plt.title("Matches Played by Each Team", fontsize=16)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Number of Matches", fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualisation 8. Matches won by each team
wins = data['winner'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(wins.index, wins.values, color='orange')
plt.title("Matches Won by Each Team", fontsize=16)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Number of Wins", fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Visualization 9 : Matches Lose by Each Team
# Create a dictionary to store the number of losses for each team
team_losses = {}

# Iterate through each row of the DataFrame
for index, row in data.iterrows():
    # Get the winning and losing teams
    winner = row['winner']
    team1 = row['team1']
    team2 = row['team2']

    # Determine the loser
    if winner == team1:
        loser = team2
    elif winner == team2:
        loser = team1
    else:  # Handle cases where there's no result ('No Result')
        continue

    # Update the losses count for the loser
    if loser in team_losses:
        team_losses[loser] += 1
    else:
        team_losses[loser] = 1

# Convert to DataFrame for better visualization
losses_df = pd.DataFrame(list(team_losses.items()), columns=['Team', 'Losses'])

# Display the team losses
print(losses_df)

# Optional: Visualization of team losses
plt.figure(figsize=(12, 6))
plt.bar(losses_df['Team'], losses_df['Losses'], color='red')
plt.title("Matches Lost by Each Team", fontsize=16)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Number of Losses", fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualisation 10. Matches with D/L method applied
dl_applied = data['dl_applied'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(dl_applied, labels=['No', 'Yes'], autopct='%1.1f%%', colors=['lime', 'red'], startangle=90)
plt.title("Matches with D/L Method Applied", fontsize=16)
plt.show()
```

```python
# Visualisation 11. Most player-of-the-match awards
top_players = data['player_of_match'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_players.index, top_players.values, color='teal')
plt.title("Top 10 Players with Most Player of the Match Awards", fontsize=16)
plt.xlabel("Number of Awards", fontsize=12)
plt.ylabel("Player", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualisation 12. Venues hosting the most matches
venues = data['venue'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(venues.index, venues.values, color='gold')
plt.title("Top 10 Venues Hosting IPL Matches", fontsize=16)
plt.xlabel("Number of Matches", fontsize=12)
plt.ylabel("Venue", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualization 13: Season Winners
season_winners = data.groupby('season')['winner'].last()
plt.figure(figsize=(12, 6))
plt.plot(season_winners.index, season_winners.values, marker='o', linestyle='-', color='purple')
plt.title('Season Winners', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Winning Team', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each point with the winner
for i, winner in enumerate(season_winners.values):
    plt.annotate(winner, (season_winners.index[i], winner), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

```python
#Visualization 14: Distribution of wins by runs

plt.figure(figsize=(10, 8))
plt.pie(team_wins, labels=team_wins.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Wins Across Teams', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Visualisation 15. Winning margin analysis
plt.figure(figsize=(10, 6))
plt.scatter(data['win_by_runs'], data['win_by_wickets'], alpha=0.7, color='magenta')
plt.title("Winning Margins (Runs vs Wickets)", fontsize=16)
plt.xlabel("Win by Runs", fontsize=12)
plt.ylabel("Win by Wickets", fontsize=12)
plt.grid(alpha=0.7)
plt.show()
```

```python
# Visualisation 16. Toss winner vs match winner
toss_match_winner = data.groupby('toss_winner')['winner'].value_counts().unstack()
toss_match_winner.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Toss Winner vs Match Winner', fontsize=16)
plt.xlabel('Toss Winner', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Visualisation 17. Umpire analysis (if umpire columns exist)
if 'umpire1' in data.columns:
    umpire1_counts = data['umpire1'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    plt.bar(umpire1_counts.index, umpire1_counts.values, color='brown')
    plt.title("Top 10 Umpires (Umpire 1)", fontsize=16)
    plt.xlabel("Umpire", fontsize=12)
    plt.ylabel("Number of Matches", fontsize=12)
    plt.xticks(rotation=90)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

if 'umpire2' in data.columns:
    umpire2_counts = data['umpire2'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    plt.bar(umpire2_counts.index, umpire2_counts.values, color='purple')
    plt.title("Top 10 Umpires (Umpire 2)", fontsize=16)
    plt.xlabel("Umpire", fontsize=12)
    plt.ylabel("Number of Matches", fontsize=12)
    plt.xticks(rotation=90)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

# Summary

This Python script performs exploratory data analysis (EDA) on an IPL (Indian Premier League) match dataset.  It begins by loading the data, cleaning it (handling missing values, converting data types, and dropping irrelevant columns), and presenting initial descriptive statistics.

The core of the script consists of 15 visualizations created using Matplotlib:

1. **Matches per Season:** Shows the number of matches played in each season.
2. **Most Successful Teams:** Displays the total number of wins for each team.
3. **Top Cities Hosting Matches:** Presents the top 10 cities hosting the most matches.
4. **Toss Decision Trends:** Illustrates the frequency of fielding and batting decisions after winning the toss (as a pie chart).
5. **Distribution of Wins by Runs:** Shows the distribution of wins based on the run margin.
6. **Distribution of Wins by Wickets:**  Shows the distribution of wins based on the wickets margin.
7. **Matches Played by Each Team:**  A bar chart showing the total number of matches played by each team.
8. **Matches Won by Each Team:** A bar chart showing the total number of matches won by each team.
9. **Matches with D/L Method:** A pie chart indicating the percentage of matches where the Duckworth-Lewis method was applied.
10. **Top Player of the Match Awards:**  A horizontal bar chart showing the top 10 players with the most "Player of the Match" awards.
11. **Venues Hosting Most Matches:** A horizontal bar chart displaying the top 10 venues with the most matches played.
12. **Winning Margins (Runs vs. Wickets):** A scatter plot showing the relationship between wins by runs and wins by wickets.
13. **Season-wise Winners:** A bar chart showing the winner of each season.
14. **Toss Winner vs. Match Winner:** A stacked bar chart analyzing the relationship between the toss winner and the match winner.
15. **Top Umpires (Umpire 1 & 2):** If the 'umpire1' and 'umpire2' columns exist in the dataset, the script generates bar charts displaying the top 10 umpires for each position.

The script provides a comprehensive visual overview of various aspects of the IPL match data, enabling insights into team performance, venue popularity, toss decisions, and more.
